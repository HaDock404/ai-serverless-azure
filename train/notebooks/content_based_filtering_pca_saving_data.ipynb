```python
import pandas as pd
import numpy as np
import os
from sklearn.metrics.pairwise import cosine_similarity
```

Agrégation des données clicks en un seul Dataframe.

```python
clicks_directory = "../data/news-portal-user-interactions-by-globocom/clicks/"
columns = ["user_id", "session_id", "session_start", "session_size", "click_article_id",
           "click_timestamp", "click_environment", "click_deviceGroup", "click_os",
           "click_country", "click_region", "click_referrer_type"]
df_clicks = pd.DataFrame(columns = columns)

for i in os.listdir(clicks_directory):
    temp_ = pd.read_csv(clicks_directory + i, index_col=None, header=0)
    liste_concat = [df_clicks, temp_]
    df_clicks = pd.concat(liste_concat, ignore_index=True)
df_clicks
```

```python
file_path_metadata = "../data/news-portal-user-interactions-by-globocom/articles_metadata.csv"
df_metadata = pd.read_csv(file_path_metadata)
df_metadata = pd.DataFrame(df_metadata)
df_metadata
```

```python
merged_df = df_clicks.merge(df_metadata, left_on='click_article_id', right_on='article_id', how='left')
merged_df
```

Pour chaque utilisateur on aggrège les différents articles et catégorie que l'utilisateur à consulté.

```python
df_grouped = merged_df.groupby('user_id').agg({
    'click_article_id': list,
    'category_id': list
}).reset_index()
df_grouped
```

Embedding des articles. On diminue la dimension par un PCA.

```python
pickle = pd.read_pickle('../data/news-portal-user-interactions-by-globocom/articles_embeddings.pickle')
pickle = pd.DataFrame(pickle)
pickle
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
pickle_scaled = scaler.fit_transform(pickle)

pca = PCA()
pca.fit(pickle_scaled)

pickle_pca = pca.transform(pickle_scaled)
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
n_components = (cumulative_variance <= 0.95).sum() + 1

pca_optimal = PCA(n_components=n_components)
pickle_pca_optimal = pca_optimal.fit_transform(pickle_scaled)
pickle_pca_optimal = pd.DataFrame(pickle_pca_optimal)
pickle = pickle_pca_optimal.reset_index().copy()
pickle
```

```python
df_temporary = df_metadata.copy()
df_temporary = df_temporary.drop(columns=["created_at_ts", "publisher_id", "words_count"])
```

```python
pickle = pickle.merge(df_temporary, left_on='index', right_on='article_id', how='left')
pickle = pickle.drop(columns=["article_id"])
pickle = pickle.rename(columns={"index": "article_id"})
cols = list(pickle.columns)
cols.insert(1, cols.pop())
pickle = pickle[cols]
pickle
```

On aggrège tous les embedding sous forme de liste dans le dataframe.

```python
df_pickle = []
df_pickle = pd.DataFrame(df_pickle, columns=["article_id", "category_id"])
df_pickle["article_id"] = pickle["article_id"].copy()
df_pickle["category_id"] = pickle["category_id"].copy()
pickle = pd.read_pickle('../data/news-portal-user-interactions-by-globocom/articles_embeddings.pickle')
pickle = pd.DataFrame(pickle)
df_pickle['embedding_concat'] = pickle.apply(lambda row: row.values.tolist(), axis=1)
df_pickle["embedding_concat"] = df_pickle["embedding_concat"].apply(lambda x: np.array(x) if isinstance(x, list) else x)
df_pickle
```

```python
final_df_pickle = '../data/csv/df_pickle.pkl'
final_df_grouped = '../data/csv/df_grouped.pkl'

df_pickle.to_pickle(final_df_pickle)
df_grouped.to_pickle(final_df_grouped)
```

Moyenne des embeddings des articles consultés par les utilisateurs pour créer un profil embeddings utilisateur.

```python
user_embeddings = {}
for user_id in df_grouped.index:
    liked_articles_embeddings = df_pickle[df_pickle['article_id'].isin(df_grouped['click_article_id'][user_id])]['embedding_concat'].values
    user_embeddings[user_id] = np.mean(liked_articles_embeddings, axis=0)
```

Comparaison entre un profil utilisateur et l'ensemble des embeddings des articles afin de déterminer les similarités entre l'utilisateur et les articles.

```python
def recommend_articles(user_id, user_embeddings, df_pickle, top_n=5):
    user_embedding = user_embeddings[user_id]
    df_similarity = df_pickle.copy()
    df_similarity['similarity'] = df_pickle['embedding_concat'].apply(lambda x: cosine_similarity([user_embedding], [x])[0][0])
    recommended_articles = df_similarity.sort_values(by='similarity', ascending=False).head(top_n)
    return recommended_articles[['article_id', 'similarity']]
```

```python
user_test = 1
recommended_articles = recommend_articles(user_test, user_embeddings, df_pickle)
print(recommended_articles)
```

```python
data_blob = []
for i in range(10):
    recommended_articles = recommend_articles(i, user_embeddings, df_pickle)
    user_id = i
    article_rec = recommended_articles['article_id'].tolist()
    data_blob.append({'user_id': user_id, 'article_rec': article_rec})

data_blob_df = pd.DataFrame(data_blob)
```
